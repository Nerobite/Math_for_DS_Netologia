Создайте случайную величину из любого выбранного вами распределения, для разных значений n сгенерируйте 1000 выборок размера n. Постройте гистрограммы средних этих выборок и сделайте выводы.

```python
exp_rv = sts.expon(5, 2)
```

```python
exp_data = exp_rv.rvs(size=1000)
```

```python
x = np.linspace(0,8,200)
pdf = exp_rv.pdf(x)
plt.hist(exp_data, density=True, bins=30, label='exp_data histogram')
plt.plot(x, pdf, label='theoretical pdf', alpha=0.5)
plt.legend()
plt.ylabel('fraction of exp_data, f(x)')
plt.xlabel('x')
plt.show()
```

```python
samples_count = 1000  # Количество повторов генерации
n1 = [2, 5, 10, 30, 50, 100]            # Размеры выборок
sample = np.zeros( (len(n1), samples_count)) #массив под выборочное среднее значение, заполненный нулями

#вычисление выборочных средних значений и занесение в массив
for i in range(len(n1)):
  for j in range(samples_count):
    sample1 = np.array(sts.expon.rvs(loc=5, scale=2, size=n1[i])) # На каждом цикле генерируем выборки указанных размеров
    sample1Mean = sample1.mean()
    sample[i, j] = sample1Mean         # Считаем среднее по каждой выборке и добавляем в массив средних

print(sample)
```
