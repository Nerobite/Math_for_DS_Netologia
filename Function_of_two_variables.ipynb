```python
x=torch.tensor(1.0, requires_grad=True)
y=torch.tensor(2.0, requires_grad=True)
f1 = 2*x**2*y**3 + 1/x + y**2*x +7
f1.backward() # вычисляем производную
print(x.grad.item())
print(y.grad.item())
```

```python
x1 = Symbol('x1')
x2 = Symbol('x2')
```

```python
fx = x2**2 - x1**2*x2 + 2*x1*x2
fx
```

```python
fig = plt.figure()
ax = fig.add_subplot(projection = '3d')

# Данные
x1 = np.linspace(-1.5, 2.5, 120)
x2 = np.linspace(-1.5, 2.5, 120)
x1, x2 = np.meshgrid(x1, x2)
Z = x2**2 - x1**2*x2 + 2*x1*x2

# График поверхности
surf = ax.plot_surface(x1, x2, Z, rstride=1, cstride=1, cmap=cm.coolwarm,  linewidth=0, antialiased=False)

# Настройка оси z
ax.set_zlim(-5, 15.0)
ax.zaxis.set_major_locator(LinearLocator(10))
ax.zaxis.set_major_formatter(FormatStrFormatter('%.02f'))


# Цветовая легенда
fig.colorbar(surf, shrink=0.5, aspect=5)

ax.azim = 15
ax.dist = 10
ax.elev = 30

plt.show()
```
