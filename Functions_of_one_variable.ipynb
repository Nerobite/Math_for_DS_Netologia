```python
y_func = sin(2*x+1)**5
y_func
```

```python
yfunc = lambda x: np.sin(2*x+1)**5
```

```python
xlist = np.linspace(-5, 5, 300)
```

```python
# Первая производная
first = y_func.diff(x)
first
```

```python
yyfunc = lambda x: 10*np.sin(2*x+1)**4*np.cos(2*x+1)
```

```python
second = first.diff(x)
second
```

```python
yyyfunc = lambda x: -20*np.sin(2*x+1)**5 + 80*np.sin(2*x+1)**3*np.cos(2*x+1)**2
```

```python
fig, ax = plt.subplots(figsize = (12, 6))
plt.plot(xlist, yfunc(xlist), label= 'Функция')
plt.plot(xlist, yyfunc(xlist), label= 'Первая производная')
plt.plot(xlist, yyyfunc(xlist), label= 'Вторая производная')
plt.xlabel('x', fontsize = 10)
plt.ylabel('y', fontsize = 10)
plt.legend()
plt.show()
```
