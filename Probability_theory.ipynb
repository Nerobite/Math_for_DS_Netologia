```python
import numpy as np
from scipy.stats import norm
import matplotlib.pyplot as plt
```

```python
fig, ax = plt.subplots(figsize = (12, 6))
#Calculate the first four moments
mean, var, skew, kurt = norm.stats(moments='mvsk')
#Display the probability density function (pdf)
x = np.linspace(norm.ppf(0.01),
                norm.ppf(0.99), 100)
ax.plot(x, norm.pdf(x),
       'r-', lw=2, alpha=0.6, label= 'Norm pdf')
ax.plot(x, norm.pdf(x, loc=1, scale= 1),
       'b-', lw=2, alpha=0.6, label= 'Norm pdf with loc=1')
ax.plot(x, norm.pdf(x, loc=2, scale= 1),
       'g-', lw=2, alpha=0.6, label= 'Norm pdf with loc=2')
ax.plot(x, norm.pdf(x, loc=3, scale= 1),
       'y-', lw=2, alpha=0.6, label= 'Norm pdf with loc=3')
plt.legend()
plt.show()

```

```python
#loc - определяет среднее значение
#scale - задает стандартное отклонение
```
